```python
from ultralytics import YOLO
from roboflow import Roboflow
```

```python
# 1. 모델 로드 (YOLO11m 모델 사용)
model = YOLO("yolo11n.pt")

# 2. 모델 학습
results = model.train(
    data="Train_BaseLine_Reduced/data.yaml", # 다운로드된 data.yaml 경로
    epochs=120,                                # 학습 반복 횟수
    imgsz=416,                                 # 이미지 사이즈
    batch=16,                                  # 배치 사이즈 (메모리 사양에 따라 조절)
    project="train_baseline_reduced_512",               # 결과 저장 프로젝트명
    name="yolo11n_train_baseline_reduced",       # 실험 이름
    device='0',
    patience=25# GPU 사용 시 0, CPU 사용 시 'cpu'
)
```

```python
model = YOLO(r"runs\detect\train_baseline_reduced_512\yolo11n_train_baseline_reduced\weights\best.pt")

metrics = model.val(
    data="test_data_new/data.yaml",
    split="val"
)
```
